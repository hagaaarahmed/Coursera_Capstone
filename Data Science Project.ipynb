# Classification 

```python
x = data_df[['WEATHER','LIGHTCOND','ROADCOND']]
x[:5]
```

```python
y = data_df[['SEVERITYCODE.1']]
y[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

# KNN Model

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
K = 15
neigh = KNeighborsClassifier(n_neighbors = K).fit(x_train,y_train)
yhat=neigh.predict(x_test)
yhat[0:5]
```

```python
print('Train Accuracy =',metrics.accuracy_score(y_train,neigh.predict(x_train)))
print('Test Accuracy =',metrics.accuracy_score(y_test,yhat))
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='entropy',max_depth=6)
tree.fit(x_train,y_train)
```

```python
predtree=tree.predict(x_test)
predtree
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=0.01,solver='liblinear').fit(x_train,y_train)
lr
```

```python
predlr=lr.predict(x_test)
predlr
```

# Model Evalution

```python
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

```python
print("KNN")
y_pred_knn=neigh.predict(x_test)
print('Jaccard of KNN=',jaccard_similarity_score(y_test,y_pred_knn))
print('F1-Score of KNN=',f1_score(y_test,y_pred_knn,average='weighted'))
print('----------------------------------------------------------------------------')
#Decision Tree
print("Decision Tree")
print('Jaccard of Descision Tree=',jaccard_similarity_score(y_test,predtree))
print('F1-Score of Descision Tree=',f1_score(y_test,predtree,average='weighted'))
print('----------------------------------------------------------------------------')
#Logistic Regression
print("Logistic Regression")
print('Jaccard of Logistic Regression=',jaccard_similarity_score(y_test,predlr))
print('F1-Score of Logistic Regression=',f1_score(y_test,predlr,average='weighted'))
y_pred_lr_proba=np.round(lr.predict_proba(x_test),-5)
print('Logloss of Logistic Regression',log_loss(y_test,y_pred_lr_proba))
print('----------------------------------------------------------------------------')
```

```python
Report=[
        {'Algorithm':'KNN',
         'Jaccard':jaccard_similarity_score(y_test,y_pred_knn),
         'F1-Score':f1_score(y_test,y_pred_knn,average='weighted'),
         'Logloss':'NA'
        },
        {'Algorithm':'DecisionTree',
         'Jaccard':jaccard_similarity_score(y_test,predtree),
         'F1-Score':f1_score(y_test,predtree,average='weighted'),
         'Logloss':'NA'
        },
        {'Algorithm':'LogisticRegression',
         'Jaccard':jaccard_similarity_score(y_test,predlr),
         'F1-Score':f1_score(y_test,predlr,average='weighted'),
         'Logloss':log_loss(y_test,y_pred_lr_proba)
        }
       ]
FinalReport=pd.DataFrame(Report)
FinalReport
```
